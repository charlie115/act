```python
import os
import sys
import pymongo
# upper_dir = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
upper_dir = '/home/info_core' # TEST
sys.path.append(upper_dir)
import pandas as pd
import json
from loggers.logger import KimpBotLogger


class InitDBClient:
    def __init__(self, host, port, user, passwd, logging_dir=None):
        self.logger = None
        self.host = host
        self.port = port
        self.username = user
        
        self.password = passwd
        self.uri = f"mongodb://{user}:{passwd}@{host}:{port}/"
        if logging_dir is not None:
            self.logger = KimpBotLogger("InitDBClient", logging_dir).logger

    def get_conn(self):
        mongo_client = pymongo.MongoClient(self.uri)
        return mongo_client
```
